```python
class attentionRegression:
    def __init__(self, w=1.0):
        """
        Initialize the attention regression model.

        Parameters:
        w (float): Positive parameter controlling attention weight distribution
        """
        self.w = w
        self.X = None
        self.y = None

    def fit(self, X, y):
        """
        'Train' the model by storing the training data.

        Parameters:
        X (numpy.ndarray): Training data matrix of shape (n, d)
        y (numpy.ndarray): Training labels of shape (n,) or (n, 1)

        Returns:
        self: The fitted estimator
        """
        # Simply store the training data for later use
        self.X = X
        self.y = y.flatten() if len(y.shape) > 1 else y

        # Return self to allow method chaining
        return self

    def predict(self, X):
        """
        Predict numerical labels for samples in X.

        Parameters:
        X (numpy.ndarray): Test samples of shape (m, d)

        Returns:
        numpy.ndarray: Predicted values of shape (m,)
        """
        if self.X is None or self.y is None:
            raise ValueError("Model has not been fitted yet. Call 'fit' first.")

        # Initialize array for predictions
        predictions = np.zeros(X.shape[0])

        # For each test point
        for i, x_hat in enumerate(X):
            # Calculate Euclidean distances
            squared_diff = (self.X - x_hat)**2
            sum_squared_diff = np.sum(squared_diff, axis=1)
            distances = np.sqrt(sum_squared_diff)

            # Calculate attention scores
            attention_scores = np.exp(-distances/self.w)

            # Normalize to probability distribution
            epsilon = 1e-12
            probabilities = attention_scores / (np.sum(attention_scores) + epsilon)

            # Calculate weighted average
            predictions[i] = np.sum(probabilities * self.y)

        return predictions



# Function to implement the model selection
def select_best_model(Xtrain, ytrain, Xvalid, yvalid, w1, w2):
   """
   Select the better model between two attentionRegression models with different w values.

   Parameters:
   Xtrain (numpy.ndarray): Training data matrix of shape (n, d)
   ytrain (numpy.ndarray): Training labels of shape (n,) or (n, 1)
   Xvalid (numpy.ndarray): Validation data matrix of shape (m, d)
   yvalid (numpy.ndarray): Validation labels of shape (m,) or (m, 1)
   w1 (float): Hyperparameter value for first model
   w2 (float): Hyperparameter value for second model

   Returns:
   tuple: (better_model, better_w, better_error)
   """
   # Initialize models
   M1 = attentionRegression(w=w1)
   M1.fit(Xtrain, ytrain)

   M2 = attentionRegression(w=w2)
   M2.fit(Xtrain, ytrain)

   # Get predictions
   y1_pred = M1.predict(Xvalid)
   y2_pred = M2.predict(Xvalid)

   # Ensure yvalid is the right shape
   yvalid_flat = yvalid.flatten() if len(yvalid.shape) > 1 else yvalid

   # Compute Mean Squared Error
   error_M1 = np.mean((yvalid_flat - y1_pred) ** 2)
   error_M2 = np.mean((yvalid_flat - y2_pred) ** 2)

   # Select better model
   if error_M1 < error_M2:
       better_model = M1
       better_w = w1
       better_error = error_M1
   else:
       better_model = M2
       better_w = w2
       better_error = error_M2

   return better_model, better_w, better_error



w1, w2 = 0.1, 1.0
best_model, best_w, best_error = select_best_model(Xtrain, ytrain, Xtest, ytest, w1, w2)
print(f"Best model has w={best_w} with validation MSE: {best_error}")
```

**Logistic Regression (customized)**

I worked on learning rate hyperparameters. I was comparing how the model behaves if we switch from a 0.1 learning rate to 1. The best model with the smaller mean square error turned out to be the one with a higher learning rate. This means that weights were adjusting to my model faster, enabling my algorithm to achieve smaller mean square error faster. However, we need to be careful, as the algorithm, in this case, might not actually learn faster when it overshoots the correct prediction.

```python
class LogisticRegressionGDwithRegularization(object):
    """Logistic Regression Classifier using gradient descent.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.


    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    loss_ : list
      Logistic loss function value in each epoch.

    """
    def __init__(self, eta=0.05, n_iter=100, alpha=0.1, random_state=2501):
        self.eta = eta
        self.n_iter = n_iter
        self.alpha = alpha  # Added parameter for L2 regularization
        self.random_state = random_state

    def fit(self, X, y, w=None):
        """ Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.

        Returns
        -------
        self : object

        """
        if w is None:
            rgen = np.random.RandomState(self.random_state)
            self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        else:
            self.w_ = w
        # rgen = np.random.RandomState(self.random_state)
        # self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.loss_ = []

        for i in range(self.n_iter):
            net_input = self.net_input(X)
            output = self.activation(net_input)
            errors = (y - output)
            self.w_[1:] += self.eta * (X.T.dot(errors) - self.alpha * self.w_[1:]) # I added L2 penalty for weights meaning that I give a little looseness for my weights to prevent overfitting
            self.w_[0] += self.eta * errors.sum()

            # Clip the values of 'output' to avoid divide-by-zero errors
            output = np.clip(output, 1e-10, 1 - 1e-10)

            # compute the logistic `loss`
            loss = -y.dot(np.log(output)) - ((1 - y).dot(np.log(1 - output))) + (self.alpha / 2) * np.sum(self.w_[1:] ** 2) # Thanks to adding L2 Regularization, I become less susceptible to overfitting giving my model a little "uncertainty"
            self.loss_.append(loss)
        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, z):
        """Compute logistic sigmoid activation"""
        return 1. / (1. + np.exp(-np.clip(z, -250, 250)))

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, 0)
        # equivalent to:
        # return np.where(self.activation(self.net_input(X)) >= 0.5, 1, 0)
```

As we can see, loss increases with alpha, meaning that the model fits worse (more regularization = simpler but worse model). I added polynomial features to my model from the sklearn library. With increasing polynomial features, I can observe a significant drop in the loss function because these features allow the model to fit better. However, these models might actually not be the best as they can overfit to the training dataset. Also, these models are unstable under regularization and may collapse. We need to be careful with setting up too big regularization strength and polynomial features.

```python
# MLP
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

scaler = StandardScaler()
Xtrain_scaled = scaler.fit_transform(Xtrain)
Xtest_scaled = scaler.transform(Xtest)

hidden_configs = [
    (5,), (10,), (20,), (30,),
    (5, 5), (10, 5), (10, 10), (20, 10), (30, 10),
    (20, 20), (30, 30), (40, 20), (40, 40), (50, 30)
]
alphas = [0.00001, 0.0001, 0.001, 0.01, 0.1]
activations = ['relu', 'tanh']

best_model_text = ""
curr_best_acc = 0
for hidden in hidden_configs:
    for alpha in alphas:
        for activation in activations:
            mlp = MLPClassifier(hidden_layer_sizes=hidden,
                                activation=activation,
                                alpha=alpha,
                                max_iter=3000,
                                early_stopping=True,
                                random_state=42)
            mlp.fit(Xtrain_scaled, ytrain)
            acc = accuracy_score(ytest, mlp.predict(Xtest_scaled))
            print(f"MLP hidden={hidden}, alpha={alpha}, activation={activation} -> Accuracy: {acc:.4f}")
            if acc > curr_best_acc:
              curr_best_acc = acc
              best_model_text = f"Best model: MLP hidden={hidden}, alpha={alpha}, activation={activation} -> Accuracy: {acc:.4f}"
print(best_model_text)
```

I was changing the following parameters:

The best model is the one with 67% accuracy. The best accuracy is achieved with two hidden layers of moderate size (20, 20), relu activation, and alpha=0.00001. Moreover, activation function 'relu' generally outperformed 'tanh'; however, it's not always the best. We need to test these activations to be able to achieve the best possible model with the highest accuracy score.
MLP model is very sensitive to activation and layer sizes. Shallow models underfit to results while deep models need careful alpha and activation adjustments

```python
# xgboost

from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score


ytrain = ytrain.ravel()
ytest = ytest.ravel()

max_depths = [3, 5, 7]
learning_rates = [0.01, 0.05, 0.1, 0.3]
n_estimators_list = [100, 200]
subsamples = [0.8, 1.0]
colsample_bytrees = [0.8, 1.0]
gammas = [0, 1]

best_model_text = ""
curr_best_acc = 0

for depth in max_depths:
    for lr in learning_rates:
        for n_estimators in n_estimators_list:
            for subsample in subsamples:
                for colsample in colsample_bytrees:
                    for gamma in gammas:
                        xgb = XGBClassifier(
                            objective='multi:softmax',
                            num_class=3,
                            eval_metric='mlogloss',
                            max_depth=depth,
                            learning_rate=lr,
                            n_estimators=n_estimators,
                            subsample=subsample,
                            colsample_bytree=colsample,
                            gamma=gamma,
                            random_state=42
                        )
                        xgb.fit(Xtrain, ytrain)
                        y_pred = xgb.predict(Xtest)
                        acc = accuracy_score(ytest, y_pred)
                        print(f"XGB depth={depth}, lr={lr}, n={n_estimators}, sub={subsample}, col={colsample}, gamma={gamma} → Accuracy: {acc:.4f}")
                        if acc > curr_best_acc:
                          curr_best_acc = acc
                          best_model_text = f"Best model: XGB depth={depth}, lr={lr}, n={n_estimators}, sub={subsample}, col={colsample}, gamma={gamma} → Accuracy: {acc:.4f}"
print(best_model_text)
```
